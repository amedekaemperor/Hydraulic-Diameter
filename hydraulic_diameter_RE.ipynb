```python
class Hydraulic_diameter_cal:
  """
  Used to determine if larminar flow or turbulent flow is present in a piple
  or duct.
  """

  def Dh_circular_tube(diameter):
    return diameter

  def Dh_square_duct(length):
    return length

  def Dh_rectangular_duct(height, width):
    return (2*height*width)/(height+width)

  def Dh_channel(height, width):
    return (4*height*width)/(2*height + width)


class Reynolds_number_cal(Hydraulic_diameter_cal):

  def __init__(self, density, velocity, viscosity):
    self.density = density
    self.viscosity = viscosity
    self.velocity = velocity
    self.temp = (self.density*self.velocity) / self.viscosity

  def circular(self, diameter):
    re = self.temp * Hydraulic_diameter_cal.Dh_circular_tube(diameter)
    if re <= 2300: return (re, 'laminar')
    elif 2300 <= re <= 10000: return (re, 'trasitional')
    else: return (re, 'turbulent')

  def square(self, length):
    return self.temp * Hydraulic_diameter_cal.Dh_square_duct(length)

  def rectangular(self, height, width):
    return self.temp * Hydraulic_diameter_cal.Dh_rectangular(height, width)

  def channel(self, height, width):
    return self.temp * Hydraulic_diameter_cal.Dh_channel(height, width)


```

```python
Re = Reynolds_number_cal(density=998, velocity=12, viscosity=0.0012)
Re.circular(diameter=3)
```
